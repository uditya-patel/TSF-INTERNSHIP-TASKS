### EDA with Data Visualisation

***Box Plot for checking Outliers***

```python
sns.boxplot(data = df, orient = 'h')
#Checking for any outliers in the dataframe
```

```python
sns.boxplot(df.Sales)
```

```python
sns.boxplot(df.Profit)
```

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
#Applying the IQR method to remove outliers
```

```python
df = df[~((df < (Q1 - 1.5*IQR))|(df > (Q3 + 1.5*IQR))).any(axis = 1)]
df
```

```python
df.shape
# Shape of the dataframe after removig outliers
```

```python
sns.boxplot(data = df, orient='h')
# The outliers have been removed now
```

```python
df.corr()
#Correlation b/w different features of the dataframe
```

```python
plt.subplots(figsize = (12,7))
sns.heatmap(df.corr(), annot = True)
#Plotting the Heatmap of different features
```

```python
ps = df.groupby('State')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
ps[:].plot.bar(color=['skyblue','red'],figsize=(15,8))
plt.title('Profit/loss & Sales across states')
plt.xlabel('States')
plt.ylabel('Profit/loss & Sales')
plt.show()
```

```python
df.groupby('Category')[['Profit','Sales']].sum().plot.bar(color=['yellow','pink'],alpha=0.9,figsize=(8,5))
plt.ylabel('Profit/loss and Sales')
plt.show()
```

```python
ps = df.groupby('Sub-Category')[['Sales','Profit']].mean().sort_values(by='Sales',ascending=False)
ps[:].plot.bar(color=['red','purple'],figsize=(15,8))
plt.title('Profit/loss & Sales across states')
plt.xlabel('Sub-Category')
plt.ylabel('Profit/loss & Sales')
plt.show()
```

***Copiers Sub-category are having highest Sales and Profit.***

***Bookcases sub-category are having large  no. of sales but very low profit followed by Table and Chairs(these sub-category have very poor profit comparing to their sales).***
